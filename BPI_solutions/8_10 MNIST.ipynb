```python
# Import datasets, classifiers and performance metrics
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
```

### load data and split into train, valid, and test sets

```python
# Load data from https://www.openml.org/d/554
X, y = datasets.fetch_openml('mnist_784', version=1, return_X_y=True)
```

Using train-test split just to get a smaller X with stratification

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=42, stratify=y)
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
Xred = tsne.fit_transform(X_test)
```

```python
plt.scatter(Xred[:,0],Xred[:,1],marker='.', c=y_test.astype(int), cmap='viridis')
plt.gca().set_aspect('equal')
```
